```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from pprint import pprint as pp
```

```python
sns.get_dataset_names()
```

```python
df_tips = sns.load_dataset('tips')
```

```python
pp(df_tips.columns)
pp(df_tips.index)
pp(df_tips.values)
pp(df_tips.keys())
pp(df_tips.shape)
```

```python
figure = plt.figure()
grid = figure.add_subplot()
# grid = figure.add_subplot(1,1,1)

# figure.clear()

grid.hist(df_tips['total_bill'], bins=10)
grid.set_title('- Histogram of Total Bill -')
grid.set_xlabel('Total Bill')
grid.set_ylabel('Frequency')

figure.tight_layout()
```

```python
figure = plt.figure()

grid = figure.add_subplot()
# grid = figure.add_subplot(1,1,1)

grid.scatter(df_tips['total_bill'], df_tips['tip'])
grid.set_title('- Scatter Plot of Total Bill vs Tips -')
grid.set_xlabel('total bill')
grid.set_ylabel('tips')

figure.tight_layout()
```

```python
df_tips.columns
```

```python
figure = plt.figure()

# grid = figure.add_subplot(1,1,1)
grid = figure.add_subplot()

grid.boxplot(
    [ 
        df_tips[ df_tips['sex'] == 'Female']['tip'],
        df_tips[ df_tips['sex'] == 'Male']['tip'] 
    ], labels=['Female', 'Male'])

grid.set_title('- boxplot of tips by sex -')
grid.set_xlabel('sex')
grid.set_ylabel('tip')

figure.tight_layout()
```

```python
def encode_sex(sex):
    print('- encode_sex({}) invoked.'.format(sex))
    
    if sex == 'Female':
        return 0
    else:
        return 1
    
    pass # encode_sex
```

```python
# df_tips['sex'].apply(encode_sex)
df_tips['sex_color'] = df_tips['sex'].apply(encode_sex)
```

```python
df_tips
```

```python
figure = plt.figure()

# grid = figure.add_subplot(1,1,1)
grid = figure.add_subplot()

grid.scatter(
    x = df_tips['total_bill'],
    y = df_tips['tip'],
    s = df_tips['size'] * 10,
    c = df_tips['sex_color'],
    alpha = .5
)

grid.set_title('- Total Bill vs Tip colored by sex and sized by size -')
grid.set_xlabel('total bill')
grid.set_ylabel('tip')

figure.tight_layout()

figure.legend()   # XX
# figure.legend(
#     handles = (df_tips['total_bill'], df_tips['tip']),
#     labels = ('total_bill', 'tip'),
#     loc='lower right'
# )

# figure.clear()
```
